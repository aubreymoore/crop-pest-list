```python
import re
from bs4 import BeautifulSoup
```

```python
# Read HTML into a string
with open('list.html') as f:
    content = f.read()
```

```python
# Use Beautiful Soup to remove HTML tags
soup = BeautifulSoup(content, "lxml")
content = soup.get_text()

# Convert to a list of strings
content = content.splitlines()
```

Direct examination of the **list.html** indicates that the data we want is contained in a block of lines where:

- the first line in the block is the last line in the document which contains the word **Agave**
- the last line in the block is the last line in the document which contains **M=x**

```python
for i, s in enumerate(content):
    if 'Agave' in s:
        first_line = i
    if 'M=x' in s:
        last_line = i

# Select the slice we want

content = content[first_line:last_line+1]
```

```python
line_type = []
for s in content:
    if s == '':
        line_type.append('blank')
    elif '[edit]' in s:
        line_type.append('crop')
    elif ':' in s:
        line_type.append('pest_order_family')
    elif ('=' in s) and (len(s) < 4):
        line_type.append('island')
    elif '(' in s:
        line_type.append('pest_scientific_name')
    else:
        line_type.append('pest_common_name')
len(content), len(line_type)
```
